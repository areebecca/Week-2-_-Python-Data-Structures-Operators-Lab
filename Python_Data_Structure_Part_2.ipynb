Bitwise Operators(Bitwise Operation)
Bitwise operations directly manipulate individual bits (s ands) within binary data, offering high-speed, low-level control for tasks like memory optimization, cryptography, and embedded system programming. They are significantly faster than standard arithmetic, working directly on integer representations.


Bitwise operators( & | ^ ~ << >>)

```python
x=10
y=3
```

```python
print(bin(10))
```

```python
print(bin(3))
```

```python
# AND (&),Sets bit to 1 if both bits are 1
x & y
#bin(10) = 1010
#bin(3)  = 0011
```

```python
# XOR(^), Sets bit to 1 if bits are different
x^y
# after it has gotten the difference in the bits and set it to one, it converts it to decimal
# 1 0 0 1 = (1×8) + (0×4) + (0×2) + (1×1)
#9
```

```python
# OR(|),sets a bit to 1 if at least one of the bits is 1
x|y
#bin(10) → 1010
#bin(3)→   0011
#(x|y)=convert(1 0 1 1) to decimal
#1011 = (1×8) + (0×4) + (1×2) + (1×1)
#11
```

```python
#Left Shift(<<),Shifts bits left
x<<2
#bin(10)=1010
# We begin to add one zero bits to the right one at a time depending on the number of shifts we should make
# First left shift= 10100
#Second Left shift= 101000
#Then convert the Last bin(101000) to decimal= (0×128) + (0×64) + (1×32) + (0×16) + (1×8) + (0×4) + (0×2) + (0×1)
```

```python
#Right Shift(>>),Shifts bits right
x>>2
#bin(10)=1010
# We begin to drop one bit at a time from the right one at a time depending on the number of shifts we should make
# First Right shift= 101
#Second Right sift=  10

```

```python
# NOT (~),Inverts all bits
~x
```

Membership Operators

Python has two membership operators:

    in
    not in

They check whether a value exists in a sequence or collection.

Sequences include:

    strings,lists,tuples,sets,dictionaries (keys)

```python
x =[2,3,4,5]
print(2 in x)
```

```python
print(0 in x)
```

STRINGS AND CONTROL FLOW

```python
name= 'RebeccA'
name
```

```python
# Modify by changing everything to upper case, upper()
name= name.upper()
name
```

```python
# Modify by changing everything to lower case,lower()
name= name.lower()
name
```

```python
# isalpha, checking if everything is alphabet.If there is any space, it makes it not an alpha
name= name.isalpha()
name
```

```python
#checking if everything is digit, isdigit()
number= '2356'
number= number.isdigit()
number
```

```python
#islower(), checking if the value(everything) is in lower case
name= 'Rebecca'
name= name.islower()
name
```

```python
#isupper(), checking every string attached is in upper case
name= 'Rebecca'
name= name.isupper()
name
```

```python
#split(), how to split string
my_world= 'Mummy you are my everything'
my_world= my_world.split()
my_world
```

```python
#join()
my_world= "_".join(my_world)
my_world
```

```python
# startswith, to check what a string/ value start with
my_world.startswith('you')
```

```python
# endswith, to check what a string / value start with
my_world= 'Mummy you are my love'
my_world.endswith('everything')
```
